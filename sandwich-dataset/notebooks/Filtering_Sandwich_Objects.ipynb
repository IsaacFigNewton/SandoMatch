```python
import pandas as pd
from google.colab import files

# Load the JSON file
data = pd.read_json("https://raw.githubusercontent.com/IsaacFigNewton/SandoMatch/refs/heads/main/sandwich-dataset/recipe-ingredients.json")

# List of terms to exclude (non-sandwich ingredients/dishes)
exclude_terms = [
    'pasta', 'rice', 'tortillas', 'wrap', 'pizza', 'soup', 'stew', 'casserole',
    'lasagna', 'stir fry', 'dumpling', 'pancake', 'taco', 'burrito', 'quesadilla',
    'curry', 'pie', 'salad', 'fajita', 'baking', 'roasting',
    'boiling', 'noodles', 'potatoes', 'lentils', 'quinoa',
    'beans', 'broth', 'stock', 'chips', 'milk', 'cream', 'garlic',
    'noodles', 'crumb', 'nuts', 'pasta', 'penne', 'vinegar', 'wine',
    'balsamic', 'sausage', 'seafood', 'clams', 'sardines', 'snails', 'beer',
    'wine', 'pudding', 'raisins', 'whipped cream', 'pecans', 'vanilla',
    'caramel', 'cinnamon', 'nutmeg', 'pie', 'salad', 'pizza', 'casserole',
    'gratin', 'soup', 'stew', 'frittata', 'cream', 'crème fraîche', 'capers',
    'rum', 'crust', 'flour'
]

# Function to filter out non-sandwich dishes
def filter_sandwiches(ingredients):
    # Check if "bread" is present and exclude recipes with unwanted terms
    likely_sandwich = "bread" in ingredients\
                              and not any(term in ' '.join(ingredients).lower() for term in exclude_terms)\
                              and len(ingredients) > 3

    return likely_sandwich

# Apply the filter function to the ingredients column
filtered_data = data[data['ingredients'].apply(filter_sandwiches)]

# Show the filtered results
filtered_data.head()
```

```python
# Convert the filtered DataFrame to JSON
json_data = filtered_data.to_json(orient='records',
                                  indent=4,
                                  lines=True)
# Save the JSON data to a file
with open('filtered_data.json', 'w') as f:
  f.write(json_data)
# Download the file
files.download('filtered_data.json')
```
