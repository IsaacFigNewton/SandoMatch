```python
import random
import re
import io
import pandas as pd
import json
import requests
from google.colab import files
import ipywidgets as widgets
from IPython.display import display

# Create a FileUpload widget that accepts JSON files
uploader = widgets.FileUpload(
    accept='.csv',
    multiple=False
)

# Display the uploader widget
display(uploader)

def handle_upload(change):
    # Declare global vars because apparently I can't use ones in the broader scope
    global ingredients_data, uploaded_filename
    if uploader.value:
        # Get the uploaded file content
        uploaded_file = next(iter(uploader.value.values()))
        uploaded_filename = uploaded_file['metadata']['name']

        # Decode the file content from bytes to a string
        csv_str = uploaded_file['content'].decode('utf-8')
        print(f"Successfully uploaded {uploaded_filename}")

        ingredients_data = pd.read_csv(io.StringIO(csv_str))


# Observe changes in the uploader widget
uploader.observe(handle_upload, names='value')
```

```python
def convert_to_json(df):
  result = {
    "Breads": dict(),
    "Meats": dict(),
    "Vegetables": dict(),
    "Cheeses": dict(),
    "Condiments": dict()
  }

  for _, row in df.iterrows():
      category = row['category']
      ingredient = row['ingredient']

      print(f"Category: {category}\
            \nIngredient: {ingredient}")

      # if the ingredient doesn't exist
      if ingredient not in result[category].keys():
          result[category][ingredient] = dict()

      if 'cost' not in result[category][ingredient].keys():
          result[category][ingredient]['cost'] = row['cost']

      # defaults to null if no calorie count provided
      if 'calories' not in result[category][ingredient].keys():
          result[category][ingredient]['calories'] = row.get('calories')

  return result
```

```python
json_data = convert_to_json(ingredients_data)

output_file = 'restructured_' + uploaded_filename.split('.')[0] + '.json'
with open(output_file, 'w') as outfile:
    json.dump(json_data, outfile, indent=4)

# Download the reformatted JSON file
files.download(output_file)
```
