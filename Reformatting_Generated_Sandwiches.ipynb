```python
# Define the set of words to remove
filter_words = {"ground", "chopped", "sliced", "freshly", "fresh"}
```

#Important Functions

```python
def get_ingredient_subtype(ingredient):
    ingredient_tags = ingredient.split(" ")
    ingr_subtype = " ".join(ingredient_tags[:-1])

    print("Ingredient Subtype: '" + ingr_subtype + "'")
    return ingr_subtype
```

```python
def add_ingredient_to_entry(entry,
                            ingredient,
                            ingredient_category,
                            exclusion_set1="",
                            exclusion_set2="",
                            exclusion_words=set()):
  for ingred_type in all_ingredients[ingredient_category].keys():
      if ingred_type in ingredient:

          subtype = get_ingredient_subtype(ingredient)
          if not subtype is None:

              exclusion_set = exclusion_words
              if len(exclusion_set1) > 0 and len(exclusion_set2) > 0:
                exclusion_set = exclusion_words\
                                .union(set(all_ingredients[exclusion_set1].keys())\
                                .union(set(all_ingredients[exclusion_set2].keys())))

              # handle the pepper case
              # if there aren't any overlapping sets of ingredients specified
              #   or if there are, but the ingredient isn't in there
              if (len(exclusion_set) <= 0)\
                  or (not ingredient in exclusion_set
                      and all(not word in exclusion_words for word in ingredient.split(" "))):

                # if the ingredient category is missing this ingredient type, add it
                if not ingred_type in entry["ingredients"][ingredient_category].keys():
                    entry["ingredients"][ingredient_category][ingred_type] = list()
                entry["ingredients"][ingredient_category][ingred_type].append(subtype)

          break
```

```python
# Define categories for organizing ingredients
def categorize_ingredients(ingredients_list):
    # entry = {
    #     "ingredients": {
    #       "bread": {bread: list() for bread in breads},
    #       "meats": {meat: list() for meat in meats},
    #       "cheeses": {cheese: list() for cheese in cheeses},
    #       "vegetables": {veggie: list() for veggie in vegetables},
    #       "condiments": {condiment: list() for condiment in condiments},
    #       "spices": {spice: list() for spice in spices}
    #     }
    #     "cost": 0,
    #     "calories": 0,
    #     "rating": random.randint(1, 5)
    # }
    entry = {
        "ingredients": {
          "bread": dict(),
          "meats": dict(),
          "cheeses": dict(),
          "vegetables": dict(),
          "condiments": dict(),
          "spices": dict()
        },
        "cost": 0,
        "calories": 0,
        "rating": random.randint(1, 5)
    }

    # map the ingredients to the correct format
    #   and filter to just the valid ingredients
    for ingredient in ingredients_list:
        # Join words in the set with | to create the regex pattern
        pattern = r"\b(" + "|".join(map(re.escape, filter_words)) + r")\b"

        # Use re.sub to remove any words in the filter_words set
        ingredient = re.sub(pattern, "", ingredient.lower()).strip()


        print("Ingredient: '" + ingredient + "'")

        add_ingredient_to_entry(entry, ingredient, "breads")
        add_ingredient_to_entry(entry, ingredient, "meats")
        add_ingredient_to_entry(entry, ingredient, "cheeses")
        add_ingredient_to_entry(entry, ingredient, "vegetables",
                                exclusion_set1="condiments",
                                exclusion_set2="spices",
                                exclusion_words={"oil"})
        add_ingredient_to_entry(entry, ingredient, "condiments")
        add_ingredient_to_entry(entry, ingredient, "spices")


    # once the ingredients list has been determined
    print(entry["ingredients"].keys())
    for category in entry["ingredients"].keys():

      print(f"Category: {category}")
      ingredients_in_category = entry["ingredients"][category]
      for ingredient_type in ingredients_in_category.keys():

        print(f"Ingredient Type: {ingredient_type}")
        for ingredient_subtype in ingredients_in_category[ingredient_type]:

          print(f"Ingredient Subtype: '{ingredient_subtype}'")

          lookup_key = ingredient_type
          if ingredient_subtype != "":
            lookup_key = ingredient_subtype + " " + lookup_key
          print(f"Ingredient being searched: '{lookup_key}'")

          try:
            ingredient = all_ingredients[category][lookup_key]
            print(ingredient)

            # estimate the sandwich's total cost
            entry["cost"] += ingredient["cost"]
            # estimate the sandwich's total calorie count
            entry["calories"] += ingredient["calories"]

          except Exception as e:
            print(f"Exception: {e}")
            continue

    print(f"Total Cost: {entry['cost']}")
    print(f"Total Calories: {entry['calories']}")
    print(f"Rating: {entry['rating']}")
    print(entry)
    return entry
```

#Main

```python
# Reformat each sandwich entry
formatted_sandwiches = []
for sandwich in sandwiches_data:
    print("Parsing sandwich...")
    print(f"ID: {sandwich['id']}")
    print(f"Cuisine: {sandwich.get('cuisine')}")
    formatted_entry = {
        "id": sandwich["id"],
        "cuisine": sandwich.get("cuisine")
    }
    formatted_entry.update(categorize_ingredients(sandwich["ingredients"]))
    formatted_sandwiches.append(formatted_entry)
    print()
    print()
    print()

# Convert to a pandas DataFrame and export
output_file = 'formatted_sandwich_entries.json'
with open(output_file, 'w') as outfile:
    json.dump(formatted_sandwiches, outfile, indent=4)

# Step 4: Download the reformatted JSON file
files.download(output_file)
```
